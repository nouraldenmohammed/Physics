```python
def mixing(Vr,Vb,drop):  
    '''
    In this code we try to simulate the following:
    
    We have tow jar, the first one is red jar with volume Vr, the second one is blue jar with volume Vb.
    The expriment start by taking a drop of size d from the red jar and transfer it to the blue jar and mix it.
    Then take another drop of the mixed jar and return it to the red jar.
    We insersted to know how many red in blue and vise versa.
    --------------------------------------
    The Inputs:
    
    Vr: the volume of the red jar
    Vb: the volume of the blue jar
    drop: the drop volume.
    
    ---------------------------------------
    The result:
    
    This code print the number of blue in the red jar
    Also, print the number of red in blue jar
    
    
    The symbol:
    the  '0' value stand for red, and  '1' stad for blue
    
    
    '''
    import random 
    ## 0 standfor red, 1 stadfor blue

    R=[0]*Vr
    B=[1]*Vb


    # the drop from red jar
    d=random.sample(R,drop)
    #print(d)


    ## adding the red drop to blue jar
    R1=R.copy()[0:len(R)-len(d)]
    B1=B.copy()+d.copy()
    #print(R1)
    #print(B1)


    ##mixing

    random.shuffle(B1)

    #print(B1)
    ##the drop from blue jar


    d=random.sample(B1,drop)
    #print(d)

    ## adding the red drop to red jar
    R2=R1.copy()+d.copy()
    # B2=B1.copy()[0:len(B1)-len(d)]
    for i in d:
        B1.remove(i)

    #print(R2)
    #print(B1)


    b_in_red=R2.count(1)
    r_in_blue=B1.count(0)
    
    print("The number of blue in red jar is {} \n the number of red in blue jar is {}".format(b_in_red,r_in_blue))

    
```

```python
for (x,y,z) in [(10000,10000,100),(500,200,10),(50,10,2),(788415,8746,45),(456541,651265,78),(68451,8451,41)]:
    mixing(x,y,z)
    print()
```

### From the above the expriment, we notice that whatever the size of the jars or the drop, on the end we find the same amount of the red in blue jar and the same amount of blue in red

# 3 (i)

```python
from numpy import *

r = 6.37*1e3
to_radian = pi/180
### capetown coordiante
xc = r*sin(125*to_radian)*cos(18*to_radian)
yc = r*sin(125*to_radian)*sin(18*to_radian)
zc = r*cos(125*to_radian)
    
    
## newyork coordiante
xn = r*sin(50*to_radian)*cos(-74*to_radian)
yn = r*sin(50*to_radian)*sin(-74*to_radian)
zn = r*cos(50*to_radian)
    
p1 = array([xc,yc,zc])
    
p2 = array([xn,yn,zn])
    ### minmum distance as stright line between cape and newyork
    
    
    
s=sqrt((p1[0]-p2[0])**2+(p1[1]-p2[1])**2+(p1[2]-p2[2])**2)
    
    
## the angle from the center of the earth between cape town and newyork
    
#alpha = 2*arcsin(s/(2*r))
alpha=arccos(p1.dot(p2.T)/(linalg.norm(p1)*linalg.norm(p2)))
    
    
## the direction of the vector between cape and newyork
    
    
direction =  90- ((pi-alpha)/2)*180/pi
    
    
    
print('''The minmum distance as stright line between cape and newyork is {} Km\n
          Cape Town is located at {} South to NewYork \n
          '''.format(s,direction))
```
