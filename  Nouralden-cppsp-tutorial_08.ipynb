```python
import numpy as np
import matplotlib.pyplot as plt
V=np.zeros((9,9))
x=[(i,j) for i in [3,4] for j in [3,4]]
for i in x:
        V[i]=9
for i in range(1,8):
    for j in range(1,8):
        if (i,j) not in x:
            V[i,j]=10
K=V.copy()
print(V)
for i in range(1,8):
        for j in range(1,8):
            if (i,j) not in x:
                V[i,j]=(V[i+1,j]+V[i-1,j]+V[i,j+1]+V[i,j-1])/4
                
while not np.allclose(V,K,rtol=1e-10):
    K=V.copy()
    for i in range(1,8):
        for j in range(1,8):
            if (i,j) not in x:
                V[i,j]=(V[i+1,j]+V[i-1,j]+V[i,j+1]+V[i,j-1])/4
    

plt.imshow(V,interpolation='gaussian')
plt.show()
np.set_printoptions(precision=5,suppress=True)
print(V)
```
